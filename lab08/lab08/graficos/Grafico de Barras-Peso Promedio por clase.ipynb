```python
# ploteando grafico de barras comparativo en pesos por cada clase
```

```python
ig, ax = plt.subplots()
pesos = [m1, m2, m3]
ano = ["Primero", "Segundo", "Tercero"]
bar_labels = ['red', 'green', 'orange']
bar_colors = ['tab:red', 'tab:green', 'tab:orange']
ax.barh(ano, pesos, label=bar_labels, color=bar_colors)
ax.set(xlabel = "Peso (lbs)", ylabel = "Curso", title = "Peso Promedio Según Año, Estudiantes Zamorano")
plt.show()
```
